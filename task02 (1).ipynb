IMPORT LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

LOAD DATASET

```python
df = pd.read_csv("C:/Users/varsh/Downloads/train.csv")
```

```python
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
#Data Cleaning
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#eda
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Class")
plt.show()
```

```python
plt.hist(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

## Key Findings
 Maleses had a higher survival rate compared  Femaleses.
- 1st class passengers had better survival chances.
- Younger passengers showed higher survival probability.
- Most passengers were between 20–40 years of age.
